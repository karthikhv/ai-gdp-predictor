```python
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1
df_clean = df[(df['Value'] >= (Q1 - 1.5 * IQR)) & (df['Value'] <= (Q3 + 1.5 * IQR))]
```

```python
df_clean = df_clean.sort_values(['CountryCode', 'ActivityCode', 'Year'])
df_clean['Lag_1'] = df_clean.groupby(['CountryCode', 'ActivityCode'])['Value'].shift(1)
df_clean['Growth_Rate'] = df_clean['Value'] - df_clean['Lag_1']
```

```python
df_clean['Target_Class'] = pd.qcut(df_clean['Value'], q=3, labels=["Low", "Medium", "High"])

df_clean.dropna().to_csv("data_cleaned.csv", index=False)
print("✅ Data Purified. 'data_cleaned.csv' created.")
```
